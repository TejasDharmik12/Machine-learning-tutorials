```python
from sklearn.datasets import load_digits
%matplotlib inline
import matplotlib.pyplot as plt
digits = load_digits()
```

```python
plt.gray()
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs',class_weight='balanced', max_iter=10000) 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data,digits.target, train_size= 0.8)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(digits.data[0:5])
```

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot = True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df['target'] = iris.target
X = iris.data
df.sample(4)
```

```python
iris.data[0]
```

```python
targets = iris.target_names
print(targets)
```

```python
X_train = df.drop(['target'], axis=1)
y_train = df['target']
```

```python
cls = LogisticRegression(solver='lbfgs',class_weight='balanced', max_iter=10000)
```

```python
cls.fit(X_train, y_train)
```

```python
X_pred = [5.1, 3.2, 1.5, 0.5]
y_pred = cls.predict([X_pred])
```

```python
print("Prediction is: {}".format(targets[y_pred]))
```
